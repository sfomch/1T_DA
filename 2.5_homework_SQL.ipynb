```python
import pandas as pd
import sqlalchemy as sa
from sqlalchemy import create_engine, URL
```

```python
params = {'host': 'localhost', 
          'port': 5432, 
          'dbname': '1t_data', 
          'user': 'postgres', 
          'password': '1'
         }
```

```python
url_object = URL.create("postgresql", 
                        username=params['user'], 
                        password=params['password'], 
                        host=params['host'], 
                        port=params['port'], 
                        database=params['dbname'])

engine = create_engine(url_object)
```

1. Вывести имена всех людей, которые есть в базе данных авиакомпаний

```python
query = ''' SELECT
                name 
            FROM 
                aero.passenger
        '''
with engine.begin() as conn:
    df = pd.read_sql_query(sa.text(query), conn)
df
```

6. Какие компании совершали перелеты на Boeing

```python
query = ''' SELECT
	            DISTINCT ac."name"
            FROM
            	aero.company ac
            JOIN 
                aero.trip atr 
            ON
	            ac.id_comp = atr.id_comp
                AND atr.plane = 'Boeing'
        '''
with engine.begin() as conn:
    df = pd.read_sql_query(sa.text(query), conn)
df
```

11. Вывести пассажиров с самым длинным именем

```python
query = ''' SELECT
	            name
            FROM
            	aero.passenger
            WHERE 
                LENGTH (name) = (
                    SELECT 
                        MAX(LENGTH (name)) 
                    FROM 
                        aero.passenger)
            '''
with engine.begin() as conn:
    df = pd.read_sql_query(sa.text(query), conn)
df
```

12. Вывести id и количество пассажиров для всех прошедших полётов

13. Вывести имена людей, у которых есть полный тёзка среди пассажиров

```python
query = ''' SELECT
	            name, count(name)
            FROM
            	aero.passenger
            GROUP BY 
                name
            HAVING      
                count(name) > 1           
        '''
with engine.begin() as conn:
    df = pd.read_sql_query(sa.text(query), conn)
df
```

14. В какие города летал Bruce Willis

```python
query = ''' SELECT
	            ap."name",
	            atr.town_to
            FROM
            	aero.passenger ap
            JOIN 
                aero.pass_in_trip pit
            ON
	            ap.id_psg = pit.id_psg
            AND 
                ap."name" = 'Bruce Willis'
            JOIN 
                aero.trip atr 
            ON
	            pit.trip_no = atr.trip_no
            '''
with engine.begin() as conn:
    df = pd.read_sql_query(sa.text(query), conn)
df
```

15. Выведите дату и время прилёта пассажира Стив Мартин (Steve Martin) в Лондон (London)

```python
query = ''' SELECT
	            time_in
            FROM
            	aero.trip atr
            JOIN 
                aero.pass_in_trip pit
            ON
                atr.trip_no = pit.trip_no 
	        JOIN 
                aero.passenger ap
            ON
	            ap.id_psg = pit.id_psg
            AND 
                ap."name" = 'Steve Martin' 
            AND 
                atr.town_to = 'London'
        '''
with engine.begin() as conn:
    df = pd.read_sql_query(sa.text(query), conn)
df
```

16. Вывести отсортированный по количеству перелетов (по убыванию) и имени (по возрастанию) список пассажиров, совершивших хотя бы 1 полет.

```python
query = ''' SELECT
	            name, 
                COUNT(name)  as count
            FROM
            	aero.passenger ap
            JOIN 
                aero.pass_in_trip pit
            ON
                ap.id_psg = pit.id_psg 
            GROUP BY 
                name
            ORDER BY count DESC, name 	       
        '''
with engine.begin() as conn:
    df = pd.read_sql_query(sa.text(query), conn)
df
```
