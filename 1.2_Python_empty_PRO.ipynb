**2. Количество отработанных часов должно быть не менее 40, а показатель производительности положительным, в случае удовлетворения необходимо посчитать эффективность работы сотрудника перемножив `время` * `производительность`, иначе же необходимо вывести замечание сотруднику**

Пример результата:

```
Введите количество отработанных часов: 42
Введите показатель производительности: 5
```

```
Введите количество отработанных часов: 38
Введите показатель производительности: 5
Количество часов должно быть больше 40
```

````python
def is_number (str_):
    try:
        float(str_)
        return True
    except:
        return False
a = input('Введите количество отработанных часов:')
while not is_number(a):
    print(f'Вы ввели не число.')
    a = input("Введите количество отработанных часов:")

b = input('Введите показатель производительности:')
while not is_number(b):
    print(f'Вы ввели не число.')
    b = input("Введите показатель производительности:")
if int(a) >= 40 and int(b) > 0:
    print(f'Эффективность работы сотрудника составляет {int(a) * int(b)} единиц в час')
else: 
    print('Плохо работаете, уважаемый')    
````

##  Цикл For

**3. Напишите программу, которая после выполнения выдает список значений, удовлетворяющих теореме Пифагора (Переберите все возможные комбинации, где сумма всех чисел будет меньше 50)**

````python
import math
for a in range(1, 50):
    for b in range(1, 50):
        c = round(math.sqrt(a**2 + b**2), 2)
        if (a + b + c) < 50:
            list_ = [a, b, c]
            print(list_)
        else:
             break
````

##  Функции

**4. Напишите функцию для поиска наименьшего общего кратного (НОК) двух чисел. Функция должна принимать на вход два параметра и возвращать наименьшее общее кратное этих чисел**

````python
def nok (a, b):
    if a < b:
        i = b
    else:
        i = a
    while True:
        if i % a == 0 and i % b == 0:
            break
        else:
            i += 1

    return i

print(nok(14,6))
        
````
